# Deep Learning Project: ANN for diabetes classification
- Diabetes Classification
- Given a dataset of various metrics can we predict if a patient has diabetes

### Step 1: Importing basic libraries

```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

### Step 2: Reading the data

```python
dataset = pd.read_csv('diabetes.csv')
dataset.head()
```

### Step 3: Checking for data quality

```python
dataset.isna().sum()
```

```python
dataset.dtypes
```

### Step 4: Creating dataset

```python
X = dataset.iloc[:,:-1]
X.head()
```

```python
y = dataset.iloc[:,-1]
y
```

### Step 5: Creating training and test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 32)
```

### Step 6: Calculating average accuracy for 10 runs

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
accuracies = []

for i in range(0,10):
    tf.random.set_seed(i)
    model = Sequential()
    model.add(Dense(units=1, activation='sigmoid', input_dim=8))
    model.compile(optimizer='adam', loss= 'binary_crossentropy', metrics=['accuracy'])
    model.fit(X_train, y_train, batch_size=100, epochs=1500, verbose=0)
    _, accuracy = model.evaluate(X_test, y_test)
    accuracies.append(accuracy)
```

```python
import matplotlib.pyplot as plt
```

```python
# Plot the accuracies with the y-axis starting from 0
plt.figure(figsize=(8, 6))
plt.plot(range(10), accuracies, marker='o')
plt.title('Accuracy Variation Across Different Random Seeds')
plt.xlabel('Iteration (Random Seed)')
plt.ylabel('Accuracy')
plt.ylim(0.5, 1)
plt.grid(True)
plt.show()
```

```python
avg_accuracy = sum(accuracies)/len(accuracies)
avg_accuracy*100
```

### Step 7: Predicting values

```python
y_pred = model.predict(X)
```

```python
y_pred = np.where(y_pred < 0.5, 0, 1)
y_pred.shape
```

### Step 8: Visualizing correct vs incorrect predictions

```python
y = y.to_numpy()
```

```python
differ = np.abs(y - y_pred.T)
differ.shape
```

```python
y_var = 'Plasma glucose concentration'
x_var = 'Age'

plt.figure(figsize=(10, 12))
scatter = plt.scatter(
    x = X[x_var], 
    y = X[y_var], 
    c=differ, 
    cmap='bwr', 
    edgecolor='k'
)

# Add labels and title
plt.xlabel(f"{x_var}")
plt.ylabel(f"{y_var}")
plt.title("Scatter Plot of Predictions vs Actual Values with Correctness")

# Add colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('Correctness (0: Incorrect, 1: Correct)')

# Display the plot
plt.show()
```
